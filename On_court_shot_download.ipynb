```python
def one_game_on_court(x):
    game_ = nbastats.loc[nbastats['GAME_ID'] == x].reset_index().drop(columns='index')
    if game_.empty:
        print(f"Warning: No rows found for GAME_ID {x}")
        return pd.DataFrame()  # Return empty DataFrame
    try:
        df = noc.players_on_court(game_)
    except Exception as e:
        print(f"ReadTimeout or other error for GAME_ID {x}: {e}")
        return pd.DataFrame()
    if df.empty:
        print(f"Warning: noc.players_on_court returned empty for GAME_ID {x}")
    return df



for game in game_ids:
    df = one_game_on_court(game)
    if not df.empty:
        on_court_stats = pd.concat([on_court_stats, df], ignore_index=True)
    else:
        print(f"Warning: No data for GAME_ID {game}")
```
